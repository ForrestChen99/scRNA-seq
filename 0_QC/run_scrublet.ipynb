```python
# 1.加载Scrublet需要的包
import scrublet as scr
import scipy.io
import matplotlib.pyplot as plt
import numpy as np
import os
import pandas as pd
```

```python
os.chdir('/home/data/t070419/publicdata/scRNA/GSE184880/')
file_to_search = "/home/data/t070419/publicdata/scRNA/GSE184880/RAW/"
filelist = os.listdir(file_to_search)
print(filelist)
dirlist=[]
for filename in os.listdir(file_to_search):
    if os.path.isdir(filename) == True:
        dirlist.append(filename)
print(dirlist)
```

```python
path = "/home/data/t070419/publicdata/scRNA/GSE184880/RAW/"
for i in filelist:
    input_dir = path + i
    counts_matrix = scipy.io.mmread(input_dir + '/matrix.mtx.gz').T.tocsc()
    counts_matrix
    out_df = pd.read_csv(input_dir + '/barcodes.tsv.gz', header=None, index_col=None, names=['barcode'])
    out_df
    scrub = scr.Scrublet(counts_matrix, expected_doublet_rate=0.06)
    doublet_scores, predicted_doublets = scrub.scrub_doublets(min_counts=2, min_cells=3, min_gene_variability_pctl=85,
                                                              n_prin_comps=30)
    # doublets占比
    print(scrub.detected_doublet_rate_)
    out_df['doublet_scores'] = doublet_scores
    out_df['predicted_doublets'] = predicted_doublets

    out_df.to_csv('/home/data/t070419/publicdata/scRNA/GSE184880/scrublet_result/' + i + 'doublet.txt',
                  index=False, header=True)
    # out_df.head()
    print(out_df["predicted_doublets"].value_counts())
```
